```python
import matplotlib.pyplot as plt
```

```python
lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15, 9.96, 7.21, 10.04, 10.2, 11.06]
```

```python
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93, 132.93, 138.92, 145.98, 148.44, 152.81]
```

```python
lime_diameter = [6.15, 7.0, 7.0, 7.69, 7.95, 7.51, 10.46, 8.72, 9.53, 10.09]
```

```python
lime_weight = [112.76, 125.16, 131.36, 132.41, 138.08, 142.55, 156.86, 158.67, 163.28, 166.74]
```

```python
plt.title('Lemons vs. Limes')
```

```python
plt.xlabel('Diameter (cm)')
```

```python
plt.ylabel('Weight (g)')
```

```python
diameter = lemon_diameter + lime_diameter
```

```python
weight = lemon_weight + lime_weight
```

```python
plt.scatter(diameter[:len(lemon_diameter)], weight[:len(lemon_weight)], color='y', label='lemons')
```

```python
plt.scatter(diameter[len(lemon_diameter):], weight[len(lemon_weight):], color='g', label='limes')
```

```python
plt.legend()
```

```python
plt.legend('y' 'g')
```

```python
plt.show()
```
