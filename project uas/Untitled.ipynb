## Explorasi Data 

```python
df.head()
```

```python
viz = df[['HARGA','LB','LT','KT','KM','GRS']]
viz.hist()
plt.show()
```

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
df['LB'].plot(kind='kde')
plt.xlabel("Luas Bangunan")

f.add_subplot(1,2,2)
sns.boxplot(df['LB'])
plt.xlabel("Luas Bangunan")

plt.show()
```

```python
f = plt.figure(figsize=(12,4))

f.add_subplot(1,2,1)
df['LT'].plot(kind='kde', color='red')
plt.xlabel("Luas Tanah")

f.add_subplot(1,2,2)
sns.boxplot(df['LT'])
plt.xlabel("Luas Tanah")

plt.show()
```

Pada bagian ini Luas Tanah sebagian besar pada angka 200 

```python

f.add_subplot(1, 2, 2)
plt.bar(df['KT'].value_counts().index, df['KT'].value_counts().values)
plt.xlabel("Kamar Tidur")
plt.ylabel("Jumlah")
plt.show()
```

```python

f.add_subplot(1, 2, 2)
sns.barplot(x=df['KM'].value_counts().index, y=df['KM'].value_counts().values)
plt.xlabel("Kamar Mandi")
plt.ylabel("Jumlah")
plt.show()
```

```python
f.add_subplot(1, 2, 2)
sns.barplot(x=df['GRS'].value_counts().index, y=df['GRS'].value_counts().values)
plt.xlabel("Garasi")
plt.ylabel("Angka")
plt.show()
```

### untuk melihat hungungan antara dipenden dan independen menggunakan pairplot

```python
f = plt.figure(figsize=(18,8))
sns.pairplot(data=df, x_vars=['LB','KT','KM','GRS'], y_vars=['HARGA'], size=5, aspect=0.75)
plt.show()
```

Semua korelasi berbentuk positif
